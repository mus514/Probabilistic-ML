```python
import jax
import jax.numpy as jnp
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
x = [-4, 4]
x_range = jnp.arange(x[0], x[1], 0.01)
normal_dist = jax.scipy.stats.norm.pdf(x_range)
uniform_dist = jax.scipy.stats.uniform.pdf(x_range, -2, 4)
laplace_dist = jax.scipy.stats.laplace.pdf(x_range, 0, 1)

plt.figure()
plt.plot(x_range, normal_dist, color='blue', linestyle='solid', label='Gaussian')
plt.plot(x_range, uniform_dist, color='green', linestyle='dashed', label='Uniform')
plt.plot(x_range, laplace_dist, color='red', linestyle='dotted', label='Uniform')

plt.legend()
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
sns.despine()
plt.show()
```

## Gaussian

```python
key = jax.random.PRNGKey(710)
key, subkey = jax.random.split(key)

x1 = jax.random.normal(key, shape=(3000, 1))
x2 = jax.random.normal(subkey, shape=(3000, 1))

plt.figure()
plt.scatter(x1, x2, marker='.', color='blue', s=4)
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.title('Gaussian')
```

## Uniform

```python
key = jax.random.PRNGKey(42)
key, subkey = jax.random.split(key)

x1 = jax.random.uniform(key, minval=-2, maxval=2, shape=(3000, 1))
x2 = jax.random.uniform(subkey, minval=-2, maxval=2, shape=(3000, 1))

plt.scatter(x1, x2, marker='.', s=1, color='green')
plt.xlim([-2.5, 2.5])
plt.ylim([-2, 2])
plt.title('Uniform')

```

## Laplace

```python
key = jax.random.PRNGKey(42)
key, subkey = jax.random.split(key)

x1 = jax.random.laplace(key, shape=(3000, 1))
x2 = jax.random.laplace(subkey, shape=(3000, 1))

plt.scatter(x1, x2, marker='.', s=1, color='red')
plt.xlim([-8, 8])
plt.ylim([-8, 8])
plt.title('Laplace')
```
