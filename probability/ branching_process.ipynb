```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pgf(t):
    return 0.3 + 0.2 * t + 0.5 * t**2

vpgf = np.vectorize(pgf)
```

$$
P_X(t) = \sum_{t \geq 0} p_k t^k
$$

```python
x = np.arange(0, 1, 0.001)
y = vpgf(x)
```

```python
plt.plot(x, y)
plt.plot(x, x)
```
